Task 1: Enhance image quality to deal with varying lighting conditions.

```python
# Necessary Library
import cv2
import numpy as np
import matplotlib.pyplot as plt

# image Load
image_path = '/content/drive/MyDrive/130C.jpg'
image = cv2.imread(image_path)

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Gaussian Blur to reduce noise
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Canny Edge Detector
edges = cv2.Canny(blurred, threshold1=50, threshold2=150)

# Optionally, overlay edges on the original image
edges_overlay = cv2.addWeighted(image, 0.8, cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR), 0.2, 0)

# Save results
cv2.imwrite('/content/drive/MyDrive/edges.jpg', edges)
cv2.imwrite('/content/drive/MyDrive/edges_overlay.jpg', edges_overlay)

# Display results
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Detected Edges')
plt.imshow(edges, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Edges Overlay')
plt.imshow(cv2.cvtColor(edges_overlay, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```

Task 2: Detect edges of vehicles and road markings to assess traffic patterns.

```python
# Load the image
image_path = '/content/drive/MyDrive/130C.jpg'
image = cv2.imread(image_path)

# Step 1: Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 2: Apply Gaussian blur to reduce noise
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Step 3: Apply Canny edge detection
edges = cv2.Canny(blurred, threshold1=50, threshold2=150)

# Save the result
cv2.imwrite('/content/drive/MyDrive/traffic_edges.jpg', edges)

# Display the results
plt.figure(figsize=(18, 10))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Detected Edges (Vehicles and Road Markings)')
plt.imshow(edges, cmap='gray')
plt.axis('off')

plt.show()
```

Task 3: Apply filters to reduce noise and improve the clarity of moving objects

```python
# Load the image
image_path = '/content/drive/MyDrive/130C.jpg'
image = cv2.imread(image_path)

# Step 1: Apply Bilateral Filtering for noise reduction while preserving edges
denoised_image = cv2.bilateralFilter(image, d=9, sigmaColor=75, sigmaSpace=75)

# Step 2: Sharpen the image using an unsharp mask
# Create a blurred version of the denoised image

blurred = cv2.GaussianBlur(denoised_image, (5, 5), 0)

# Enhance sharpness by subtracting blurred from original
sharpened = cv2.addWeighted(denoised_image, 1.5, blurred, -0.5, 0)

# Save results
cv2.imwrite('/content/drive/MyDrive/denoised_image.jpg', denoised_image)
cv2.imwrite('/content/drive/MyDrive/sharpened_image.jpg', sharpened)

# Display results
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Denoised Image')
plt.imshow(cv2.cvtColor(denoised_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Sharpened Image')
plt.imshow(cv2.cvtColor(sharpened, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```
